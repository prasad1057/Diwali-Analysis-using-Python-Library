# Exploratory Data Analysis

**Gender**

```python
df.columns
```

```python
# countplot = Show the counts of observations in each categorical bin using bars
sns.countplot(x='Gender',data =df)
```

```python
ax = sns.countplot(x='Gender',data=df,color='r')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
df.head()
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

From the above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than man

**State**

```python
df.columns
```

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sales_state
```

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(9)

sns.set(rc={'figure.figsize':(15,5)})  #(width,height)       # if u not this then it show x axis overlapped
sns.barplot(x='State', y='Orders', data=sales_state)
```

From above graph we can see that most of the orders are from Uttar Pradesh, Maharashtra and Karnataka respectively but total sales/amount is from UP, Karnataka, and hen Maharashtra

**Marital Status**

```python
df.columns
```

```python
ax2 = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax2.containers:
    ax2.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Marital_Status',y='Amount', data=sales_state, hue='Gender')
```

From above graphs we can see that most buyers are maried (women) adn they have high purchasing power

**Occupation**

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x='Occupation', data=df)

for bar in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y='Amount', data=sales_occ)
```

From above graphs we can see that most of the buyers are working in IT, Healthcare, and Aviation Sector

**Product Category**

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x='Product_Category', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_prc = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category', y='Amount', data=sales_prc)
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category


```python
sales_state = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID', y='Orders', data=sales_state)
```

```python
# top 10 sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
